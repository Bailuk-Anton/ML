```python
# 1.1 Выбор исследуемого датасета.
# https://www.kaggle.com/datasets/uom190346a/sleep-health-and-lifestyle-dataset?resource=download
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/Sleep_health_and_lifestyle_dataset.csv")
```

```python
# Разделим Blood Pressure на два числовых столбца
df[['Systolic', 'Diastolic']] = df['Blood Pressure'].str.split('/', expand=True).astype(int)
```

```python
# 1.3. Для каждого атрибута нужно определить:
# 	•	среднее значение, ско
df.describe()
```

```python
# 1.3
#	•	построить гистограмму распределения значений, определить есть ли выбросы 
# Разделим признаки

# Числовые признаки:
# Age
# Sleep Duration (hours)
# Quality of Sleep (scale: 1–10)
# Physical Activity Level (minutes/day)
# Stress Level (scale: 1–10)
# Heart Rate (bpm)
# Daily Steps
# Systolic
# Diastolic

# Категориальные признаки:
# Gender
# Occupation
# BMI Category
# Blood Pressure (систолическое/диастолическое — требует разбора!)
# Sleep Disorder
numeric_cols = [
    'Age',
    'Sleep Duration',
    'Quality of Sleep',
    'Physical Activity Level',
    'Stress Level',
    'Heart Rate',
    'Daily Steps',
    'Systolic',
    'Diastolic'
]

categorical_cols = [
    'Gender',
    'Occupation',
    'BMI Category',
    'Sleep Disorder'
]
```

```python
# 1.3
#	•	построить гистограмму распределения значений, определить есть ли выбросы 
import math

fig, axes = plt.subplots(3, 3, figsize=(16, 4 * 3))
axes = axes.flatten()

for i, col in enumerate(numeric_cols):
    # Гистограмма
    sns.histplot(df[col], kde=True, ax=axes[i], bins=20)
    axes[i].set_title(f'{col}\n(μ={df[col].mean():.1f}, σ={df[col].std():.1f})')
    
plt.tight_layout()
plt.show()
```

```python
# 1.3 определить есть ли выбросы

# Настройка стиля
sns.set(style="whitegrid")

# Построим отдельный boxplot для каждого признака
for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=df, y=col, color='lightblue')
    plt.title(f'Boxplot: {col}')
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()
    
#Вывод: выбросы есть  в столбце 'Heart Rate'
```

```python
# 1.3. Для каждого атрибута нужно определить:
# есть ли пропущенные значение, сколько
# предложить вариант обработки пропущенных значений
print(df.isnull().sum())

# Вариант обработки пропущенных значений: Изменить 'N/A' на 'None'
```

```python
df = df.fillna('None')
```

```python
# 1.4. Определить корреляцию между параметрами
# 	•	какие атрибуты высококоррелированы, определить характер корреляции 
# 	•	какие атрибуты не имеют корреляцию
# 	•	постройте графики рассеивания (предпочтительнее матрицу графиков рассеивания)
# 	•	проанализируйте полученные результаты.

# Создадим копию датасета в числовых столбцах
df_numeric = df[numeric_cols]

# Матрица корреляций (Пирсона)
# Вычисляем корреляцию
corr_matrix = df_numeric.corr()

# Визуализация тепловой картой
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
plt.title("Матрица корреляций (Пирсона)")
plt.tight_layout()
plt.show()
```

```python
# 0.00 – 0.19 Очень слабая / отсутствует
# 0.20 – 0.39 Слабая
# 0.40 – 0.59 Умеренная
# 0.60 – 0.79 Сильная
# 0.80 – 1.00 Очень сильная

# Очень сильный хар-тер корреляции: Quality of Sleep & Sleep Duration
#                                   Stress Level & Quality of Sleep(обратная корреляция)
#                                   Stress Level & Sleep Duration(обратная корреляция)

# Сильный хар-тер корреляции:       Heart Rate & Stress Level
#                                   Daily Steps & Phys. Active Level
#                                   Age & Systolic/Diastolic
#                                   Heart Rate & Quality of Sleep(обратная корреляция) 

# Умеренный хар-тер корреляции:     Age & Quality of Sleep
#                                   Stress Level & Age()
#                                   Heart Rate & Sleep Duration(обратная корреляция)

# Слабый хар-тер корреляции:        Sleep Duration & Age
#                                   Phys. Active Level & Sleep Duration
#                                   Heart Rate & Systolic/Diastolic
#                                   Heart Rate & Age(обратная корреляция)

# У остальных характеристик очень слабая корреляция или отсутствует
```

```python
# Построим график рассеивания для параметров с очень сильной корреляцией
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(18, 5))  # 1 строка, 3 столбца

# График 1: Quality of Sleep vs Sleep Duration
df.plot.scatter(x='Quality of Sleep',
                y='Sleep Duration',
                c='DarkBlue',
                ax=axes[0])
axes[0].set_title('Quality of Sleep vs Sleep Duration')

# График 2: Sleep Duration vs Stress Level
df.plot.scatter(x='Sleep Duration',
                y='Stress Level',
                c='DarkBlue',
                ax=axes[1])
axes[1].set_title('Sleep Duration vs Stress Level')

# График 3: Quality of Sleep vs Stress Level
df.plot.scatter(x='Quality of Sleep',
                y='Stress Level',
                c='DarkBlue',
                ax=axes[2])
axes[2].set_title('Quality of Sleep vs Stress Level')

plt.tight_layout()
plt.show()
```
